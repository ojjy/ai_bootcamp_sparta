## Deep Learning 스터디

- 딥러닝 : Artificial Neural Network를 기반으로 한 기계학습의 한 분야 층을 깊게 쌓는 다는 Deep, learning 데이터로 부터 학습 
- 다층 신경망을 사용하여 데이터로 부터 특징을 자동으로 학습하고 이를 통해 복잡한 문제를 해결
하나의 신경망은 어떤 특징을 배우거나 규칙을 배울수 있는 인공신경망이라고 하는 네트워크 이다. 
하나의 머신러닝의 모듈이 여러층으로 쌓여 데이터로 부터 특징을 잘학습할수 있게 하는게 deep learning 이를 바탕으로 예측 분류 생성등의 작업가능

개사진, 고양이 사진 분류 픽셀의 관계가 중요한 역할을 햇으면 딥러닝은 픽셀의 관계를 추출. 문제해결을 위해 가장 잘 동작하는 특징
입력이 들어오면 인공신경망을 통과하여 최종 출력이 나온다. 
output이 개, 고양이면 분류, 가격이면 예측 정확도가 높은 대부분의 모델은 DL로 부터 파생

비선형 추론이 가능. 선형이란 정답=a*특징+b 직선을 나타내는 방법 그러나 선형은 그다지 좋은 성능을 내지 못한다
나이 = a특징 +b 키를 맞추는 문제라고 가정하면 10*키 +5 105 1005 나이와 키간의 관계 선형이 아니라 복잡한 관계. 선형관계 복잡한 데이터 학습 불가. 비선형은 복잡한관계 학습가능
여러층의 인공신경망을 사용하는데 하나의 인공신경망보다 여러개의 층을 거치면서 추론능력을 향상시켜 데이터의 고차원특징 즉 복잡한 특징을 학습할수 있다. 
머신러닝에서는 데이터로 부터특징 찾아낸다. dl에서는 중요한 부분을 자동으로 추출. 물론 데이터를 전처리를 통해서 dl이 잘이해할수 잇는 형태로 바꾸면 성능이 올라간다. ml은 경우 중요한 특징을 골라준 반면 dl은 중요한 부분 자동으로 추출

ann, cnn 신경망을 깊게 쌓은것 - dl
대규모 데이터 복잡한 문제라 학습시간 많이 걸린다.


## 신경망의 기본원리

인공신경망:  퍼셉트론(인공신경망의 기본단위)으로 구성되어 있는 네트워크. 다중 퍼셉트론
퍼셉트론: 사람의 뇌를 모델링 한것. 사람이 생각하는 것과 비슷하게 동작. 사람은 뇌에서 뉴런(퍼셉트론)을 통해 신경망(인공신경망)을 형성하고 사고를 하는데 이것을 흉네
퍼셉트론은 여러 입력값을 받아 가중치를 더하고 이를 모두 더한 후 활성화 함수를 통해 출력값 결정
이 방식으로 사고가 가능하다. 이를 모델링한게 ann(인공신경망)
귀의 개수, 다리의 개수를 통해 사람인지 개인지 퍼셉트론을 만든다고 가정

다층퍼셉트론 - 퍼셉트론 여러층으로 쌓으면 복잡한 문제 해결 가능. 
단일퍼셉트론 xor논리회로를 표현 불가. 다층 퍼셉트론은 다선형성 학습가능. 활성화함수가 단일함수의 경우 선형레이어를 몇개 통화시키던 비선형성 사용한는 이유.

활성화함수 필요성 - 신경망 단순 선형. 복잡한 패턴 학습 불가. 
입력값을 출력값으로 변환. 활성화 함수 비선형형 도입하여 복잡한 패턴 합습가능


손실함수 - 모델의 예측과 실제 값 차이 
목적: 모델의 성능평가, 최적화 알고리즘을 통해 모델을 학습.
최적화 알고리즘 : 손실함수의 값이 줄어들도록 a, b값 업데이트 - 손실함수가 최저가 되도록 최적화 알고리즘 도입. 학습은 가중치 a, b 업데이트. 모델의 성능을 올랄갈지 찾는다.

최적화 알고리즘 종류
sgd. adam
손실함수 최소화하는 방식 여러개, 종류에 따라 파라미터 업데이트 하는 방법 달라 각자 도달할수 있는 optimum지점 상이.

역전파
